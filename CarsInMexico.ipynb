```python
import pandas as pd
```

```python
df = pd.read_csv('./coches_Mexico.csv', index_col=False) 
```

```python
df.head()
```

```python
df = df[df['Año'] > 2011]
df = df[df['Cantidad'] != '-']
df['Modelo'] = df['Modelo'].str.rstrip('-')
df['Modelo'] = df['Modelo'].str.lower()
df['Marca'] = df['Marca'].str.lower()
```

```python
df.head()
```

```python
df['Cantidad'] = pd.to_numeric(df['Cantidad'])
```

```python
df.info()
```

```python
total_cars_df = df.groupby(['Marca','Modelo']).agg({'Cantidad': 'sum'}).sort_values(by='Cantidad', ascending=False)
```

```python
total_cars_df['Porcentaje'] =  total_cars_df['Cantidad'] / total_cars_df['Cantidad'].sum()
total_cars_df['Porcentaje'] =  total_cars_df['Porcentaje'].round(decimals = 4)
```

```python
total_cars_df.head(100)
```

```python
total_cars_df.to_csv('cars_per_model.csv')
```
